## Simple NER

```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp("apple is looking at buying U.K. startup for $1 billion.")
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

## Advanced NER

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp("Google has offices in New York and London.")
```

```python
gpe_ent=[ent.text for ent in doc.ents if ent.label_ == "GPE"]
```

```python
print("Locations:",gpe_ent)
```
